```python
import numpy as np
import codecs
import urllib
import json
import cv2
import pandas as pd
import requests
import bs4
import imutils
import matplotlib
#matplotlib.use("Agg")
from os import listdir
import matplotlib.image as mpimg 
import matplotlib.pylab as plt
from matplotlib.pyplot import savefig  
%matplotlib inline
from flask import Flask, render_template, Response

class VideoCamera(object):
    def get_page_urls(url):
        json_file = []
        response = requests.get(url)
        soup = bs4.BeautifulSoup(response.text, "html5lib")
        page_info = [a.attrs.get('href') for a in soup.select('li a')]
        for i in page_info:
            if i[-4:] == 'json':
                json_file.append(i)
        return json_file

    def get_position(junction, frame):
        x = []
        y = []
        for a, i in enumerate(range(frame.shape[0])):
            y.append(int(frame[a].get(junction).get('x')))
            x.append(int(frame[a].get(junction).get('y')))
        return x, y
    
    def __init__(self):
        url = 'http://www.cadietrich.net/reports/legotracker/database/'
        json_file = get_page_urls(url)
        data_num = input("data_num = ")
        s = json_file[int(data_num)]
        read_json = urllib.request.urlopen(url+s)
        data = json.loads(read_json.read().decode('utf8'))
        frames = np.array(data[u'frames'])
        video = data[u'video']
        play_id = np.array(data[u'play_id'])
#         print(play_id)
        timestamp = np.array(data[u'timestamp'])
        junctions = ['left_ankle','left_elbow','left_hip','left_knee','left_shoulder','left_wrist',
                    'right_ankle','right_elbow','right_hip','right_knee','right_shoulder','right_wrist'
                    ]

        left_ax, left_ay = get_position('left_ankle', frames)
        left_ex, left_ey = get_position('left_elbow', frames)
        left_hx, left_hy = get_position('left_hip', frames)
        left_kx, left_ky = get_position('left_knee', frames)
        left_sx, left_sy = get_position('left_shoulder', frames)
        left_wx, left_wy = get_position('left_wrist', frames)
        right_ax, right_ay = get_position('right_ankle', frames)
        right_ex, right_ey = get_position('right_elbow', frames)
        right_hx, right_hy = get_position('right_hip', frames)
        right_kx, right_ky = get_position('right_knee', frames)
        right_sx, right_sy = get_position('right_shoulder', frames)
        right_wx, right_wy = get_position('right_wrist', frames)

        download_video = urllib.request.FancyURLopener()
        download_video.retrieve(video,'video.mp4')

        camera = cv2.VideoCapture('video.mp4')
        self.video = camera
    
    def __del__(self):
        self.video.release()
    
    def get_frame(self):
        success, image = self.video.read()
        ret, jpeg = cv2.imencode('.jpg', image)
        return jpeg.tobytes()

app = Flask(__name__)

@app.route('/')  # 主页
def index():
    return render_template('index.html')

def gen(camera):
    while True:
        frame = camera.get_frame()
        yield (b'--frame\r\n'
               b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n\r\n')

@app.route('/video_feed') 
def video_feed():
    return Response(gen(VideoCamera()),
                    mimetype='multipart/x-mixed-replace; boundary=frame')   

if __name__ == '__main__':
    app.run(host="127.0.0.1", debug=True, port=5000)
```
